# Euler's Method

We will use Euler's Method to solve a first order differential equation in an interval. Specifically,
$$\dfrac{dx}{dt} = f(x, t)$$

First, we need the relevant imports

```python
import math
import numpy as np
import pylab as pl
```

Now let's define our problem.
$$f(x, t) = -x^3 + \sin t$$
in the interval $(0, 10)$ with the initial conditions $(t_0, x_0) = (0, 0)$.

```python
f = lambda x, t: -x**3 + math.sin(t)
a = 0.0                    # Start of Interval
b = 10.0                   # End of Interval
N = 1000                   # Number of Steps
h = (b - a) / N            # Size of a Step
x = 0.0                    # Initial Condition (at (a))
```

Now let's setup our points and apply Euler's method.

```python
ts = np.arange(a, b, h)
xs = []

for t in ts:
    xs.append(x)
    x += h * f(x, t)
```

Now, we plot our function $x(t)$.

```python
pl.plot(ts, xs)
pl.xlabel("t")
pl.ylabel("x(t)")
pl.show()
```
